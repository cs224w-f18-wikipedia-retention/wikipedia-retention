```python
import snap

G = snap.LoadEdgeList(snap.PUNGraph, "../data/processed/2007-1-enwiki-projection-user.csv")
```

```python
CoreIDSzV = snap.TIntPrV()
%time kValue = snap.GetKCoreNodes(G, CoreIDSzV)
for item in CoreIDSzV:
    print "order: %d nodes: %d" % (item.GetVal1(), item.GetVal2())
```

```python
import matplotlib.pyplot as plt
import numpy as np

x, y = zip(*[(x.GetVal1(), x.GetVal2()) for x in CoreIDSzV])
ax = plt.gca()
ax.semilogy(x, y)
plt.show()
```

```python
CoreIDSzV = snap.TIntPrV()
kValue = snap.GetKCoreEdges(G, CoreIDSzV)
for item in CoreIDSzV:
    print "order: %d edges: %d" % (item.GetVal1(), item.GetVal2())
```

```python
x, y = zip(*[(x.GetVal1(), x.GetVal2()) for x in CoreIDSzV])
ax = plt.gca()
ax.semilogy(x, y)
plt.show()
```
